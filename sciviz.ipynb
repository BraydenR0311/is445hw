```python
import yt
```

```python
ds = yt.load('ArepoBullet/snapshot_150.hdf5')
```

```python
ds.derived_field_list
```

```python
p = yt.SlicePlot(ds, 'z', ('gas', 'density'))
```

```python
p.zoom(9)
```

```python
p.pan_rel((-.2,0))
```

```python
p2 = yt.SlicePlot(ds, 'x', ('gas', 'density'))
```

```python
p2
```

```python
p2.zoom(9)
```

```python
p2.annotate_velocity()
```

```python
sphere = ds.sphere('c', (100.0, 'kpc'))
plot = yt.ProfilePlot(
    sphere,
    ('gas', 'density'),
    ("gas", "temperature")
)
plot.set_log(("gas", "temperature"), False)

plot
```

## First Viz

 I took a slice out of the "ArepoBullet" galaxy cluster merger data. It is a visualization across the z axis displaying the density. It's really cool when you zoom in and you can see the merge. I didn't annotate velocity because I didn't like the look with the arrows.

## Second Viz

This is the same idea as the first viz. It's another slice of the 3D visualization, but along the x axis. Because it's a different perspective, it looks completely different. I annotated velocity with this one, and you can see the arrows pointing towards the center. Instead of being an arrow shape, like the first viz, it looks like a circular target.

## Third Viz

My third viz, from what I understand, is the mean temperature over density of a spherical region of 100 kpc within the data. ## Comparing and Contrasting

Having multiple visualizations on a single dataset is important in getting a holistic understanding of the data. For instance, it's good to represent it spacially because we humans understand things in space very easily, but to also make a chart that represents something else, like temperature or density over time, because we've evolved with abstract thinking and can understand different elements of data in this way. If we only view something spacially, we are missing what could be vital information, but is not observable to the eye.Likewise, 2D data can be meaningless if we don't know what the data looks like in space. Also, being able to view 3D data at multiple slices, like I did, is important, since we can only view it in "2D", because we aren't fourth-dimension creatures (yet...).
